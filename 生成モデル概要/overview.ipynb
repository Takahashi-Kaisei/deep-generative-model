### 生成モデルとは



生成モデルの目標

- 自分でモデリングした確率モデル．それを，この世界中にあるどこかのデータ分布(真の分布，母分布と言ってもいいかも)に近づけること．

データ分布は，この世界中のどこかにあると仮定するんだけど，それは残念ながら私たちにはわからない．\
だけど，確率モデルを作り，データ分布に近づけることで，データ分布の性質を知ろうね．というのが生成モデル．

構成論的アプローチ

データを理解したければ，データが作れるようになろう。そうすれば理解したことになる．

生成モデルだとできること
- 生成
- 密度推定
- 欠損値補完，ノイズ除去

気をつけたいのは，生成モデルには本物とのズレがあること．

データ分布と生成モデルを完全に一致させることは難しい．保証はできない．

補完と言いつつ，本物っぽく作り直してるニュアンスになる．

### 識別モデル

生成モデルの対比として，よく出てくるのが識別モデル．

|識別モデル|生成モデル|
|---|---|
|$p_x(y)$を直接データ集合から学習|同時分布$p(x, y)$をデータ集合から学習|
|学習した$p_x(y)$を用いて予測|$p_x(y) = \frac{p(x, y)}{\int p(x, y)dy}$を求めて予測|

- 同時分布(同時確率分布) $p(x, y)$
    - 2つの変数$x, y$のとりうる値と，その確率との対応関係のこと．
    - 同時ど名前がついていることからも分かるとおり，$x, y$はそれぞれの値によらない．周辺化していないということ．
- 周辺分布
    - 周辺化
- 条件付き分布
